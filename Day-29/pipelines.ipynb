```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['PassengerId' , 'Name' , 'Ticket' , 'Fare' , 'Cabin'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.impute  import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline , make_pipeline 
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test , y_train , y_test  = train_test_split(df.drop(columns=['Survived'] ), df['Survived'] , test_size=0.2 , random_state=42)
```

```python
x_train
```

```python
df['Embarked'].unique()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[5])
],remainder='passthrough')
```

```python
trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,5])
],remainder='passthrough')
```

```python
trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,9))
])
```

```python
trf5 = DecisionTreeClassifier()
```

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf5',trf5)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
