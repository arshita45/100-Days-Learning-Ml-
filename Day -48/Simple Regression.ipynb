```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = pd.read_csv("placement (1).csv")
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'] , df['package'])
plt.xlabel('cgpa')
plt.ylabel('package')
plt.show()
```

Linear regression is a supervised learning algorithm it has three types simple, mutiple and polynomial , it works on basis of finding the best fit line through our graph that is closest to all points .
It has a slope and a intercept , slope defines the weightage of input column on output column .

```python
x = df.iloc[:,0:1]
```

```python
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```
